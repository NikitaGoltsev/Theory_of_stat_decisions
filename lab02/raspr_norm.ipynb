```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
 
# Создаем нормальное распределение вероятностей среднего значения 0 и стандартного отклонения 1
distribution = norm(0, 1)
 
# Генерируем случайные значения из созданного распределения
data = distribution.rvs(size=1000)
 
# Строим гистограмму данных и нарисуем график плотности распределения
plt.hist(data, bins=30, density=True, alpha=0.7)
x = np.linspace(-4, 4, 100)
plt.plot(x, distribution.pdf(x), 'r', lw=2)
plt.xlabel('Значение')
plt.ylabel('Плотность распределения')
plt.title('Нормальное распределение')
plt.show()
```

```python
from scipy.stats import norm
 
data = [1.2, 2.5, 3.7, 4.1, 5.8, 6.3]
 
# Оцениваем параметры нормального распределения на основе набора данных
mean, std = norm.fit(data)
 
print("Среднее значение:", mean)
print("Стандартное отклонение:", std)
```
